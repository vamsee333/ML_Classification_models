```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")
```

```python
data.head()
```

```python
features = data.drop("species", axis=1)
labels = data["species"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import warnings
warnings.filterwarnings("ignore",category=FutureWarning)
warnings.filterwarnings("ignore",category=DeprecationWarning)
from sklearn.model_selection import GridSearchCV
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, labels, test_size=0.2, random_state = 42)
```

```python
model = MLPClassifier()
params={
    "activation":["relu","tanh","logistic"],
    "hidden_layer_sizes":[(10,),(20,),(5,2)],
    "learning_rate":["constant","invscaling","adaptive"]
}

cv = GridSearchCV(model, params, cv=5)
cv.fit(X_train, Y_train)
```

```python
def print_results(cv):
    print("the best parameters are {}".format(cv.best_params_))
    
    mean = cv.cv_results_["mean_test_score"]
    for mean, params in zip(mean, cv.cv_results_["params"]):
        print("mean {} - paramters are {}".format(mean, params))
```

```python
print_results(cv)
```
